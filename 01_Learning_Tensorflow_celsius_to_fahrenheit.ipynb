```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
celsius_q    = np.array([-100,  -90, -60, -40, -10,  0,    8, 15,   22,    38,  50,  60,  70,  80,  90, 100],  dtype=float)
fahrenheit_a = np.array([-148, -130, -76, -40,  14, 32, 46.4, 59, 71.6, 100.4, 122, 140, 158, 176, 194, 212],  dtype=float)
```

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
model = tf.keras.Sequential([l0])
```

```python
model.compile(loss='mean_squared_error',
               optimizer=tf.keras.optimizers.Adam(0.5))
history = model.fit(celsius_q, fahrenheit_a, epochs=120, verbose=False)
```

```python
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
model.compile(loss='mean_squared_error',
               optimizer=tf.keras.optimizers.Adam(0.01))
history = model.fit(celsius_q, fahrenheit_a, epochs=100, verbose=False)
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
model.compile(loss='mean_squared_error',
               optimizer=tf.keras.optimizers.Adam(0.001))
history = model.fit(celsius_q, fahrenheit_a, epochs=100, verbose=False)
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
print(model.predict([100]))
print(model.predict([110]))
print("These are the layer variables: {}".format(l0.get_weights()))
```

```python
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)

model = tf.keras.Sequential([l0, l1, l2])
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)

print(model.predict([100]))
```
